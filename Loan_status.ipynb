```
import pandas as pd
```

```
url = 'https://raw.githubusercontent.com/milindsinha-bitspilani/Loan_Status/master/test.csv.csv'
df1 = pd.read_csv(url)
```

```
df1
```

```
url = 'https://raw.githubusercontent.com/milindsinha-bitspilani/Loan_Status/master/train_csv.csv'
df_train = pd.read_csv(url)
```

```
df_train
```

```
df1.dtypes
```

```
df1.dropna(inplace=True)
df_train.dropna(inplace=True)
```

```
df1.replace('3+',5, inplace=True)
```

```
df1
```

```
df_train.replace('3+',5,inplace=True)
df1.replace('Rural',0.0,inplace=True)
df_train.replace('Rural',0.0,inplace=True)
df1.replace('Urban',1.0,inplace=True)
df_train.replace('Urban',1.0,inplace=True)
df1.replace('Semiurban',0.5,inplace=True)
df_train.replace('Semiurban',0.5,inplace=True)
```

```
df1
```

```
df1['Dependents'] = df1['Dependents'].apply(pd.to_numeric) 
df_train['Dependents'] = df_train['Dependents'].apply(pd.to_numeric) 
```

```
df1.dtypes
```

```
df1.replace('Yes',1,inplace=True)
df_train.replace('Yes',1,inplace=True)
df1.replace('No',0,inplace=True)
df_train.replace('No',0,inplace=True)
```

```
df1
```

```
df1.replace('Male',0,inplace=True)
df_train.replace('Male',0,inplace=True)
df1.replace('Female',1,inplace=True)
df_train.replace('Female',1,inplace=True)
```

```
df1
```

```
df_train
```

```
df_train.replace('N',0,inplace=True)
df_train.replace('Y',1,inplace=True)
```

```
df_train
```

```
df_train.replace('Graduate',1,inplace=True)
df_train.replace('Not Graduate',0,inplace=True)
df1.replace('Graduate',1,inplace=True)
df1.replace('Not Graduate',0,inplace=True)
```

```
df_train.corr(method='pearson')['Loan_Status'].sort_values()
```
